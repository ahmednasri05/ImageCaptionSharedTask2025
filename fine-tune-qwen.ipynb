```python
import os
import shutil

INPUT_DIR = "/kaggle/input/imageeval-emnlp"
WORKING_DIR = "/kaggle/working"
TRAIN_DIR = os.path.join(WORKING_DIR, "Train")
IMAGES_DIR = os.path.join(TRAIN_DIR, "images")

# Make sure output directories exist
os.makedirs(IMAGES_DIR, exist_ok=True)

# Copy Excel file
count = 0
for subfolder in os.listdir(INPUT_DIR):
    outer_path = os.path.join(INPUT_DIR, subfolder)
    inner_path = os.path.join(outer_path, "ImageEvalTrain")  # New fix

    if os.path.isdir(inner_path):
        for file in os.listdir(inner_path):
            if file.lower().endswith((".jpg", ".jpeg", ".png")):
                src = os.path.join(inner_path, file)
                dst = os.path.join(IMAGES_DIR, file)  # Use original name
                shutil.copy(src, dst)
                count += 1

print(f"✅ Copied {count} images to {IMAGES_DIR}")
```

```python
import torch

def list_available_gpus():
    print(f"Total GPUs Available: {torch.cuda.device_count()}")
    for i in range(torch.cuda.device_count()):
        gpu_name = torch.cuda.get_device_name(i)
        total_mem = torch.cuda.get_device_properties(i).total_memory / 1024**3
        reserved_mem = torch.cuda.memory_reserved(i) / 1024**3
        allocated_mem = torch.cuda.memory_allocated(i) / 1024**3
        print(f"GPU {i}: {gpu_name}")
        print(f"  Total Memory:     {total_mem:.2f} GB")
        print(f"  Reserved Memory:  {reserved_mem:.2f} GB")
        print(f"  Allocated Memory: {allocated_mem:.2f} GB\n")

if torch.cuda.is_available():
    list_available_gpus()
else:
    print("❌ No CUDA-compatible GPU available.")
```
